```python
import pymongo
import pandas as pd
import numpy as np
```

```python
# read data and sorted by event_time
raw_data = pd.read_csv('/Users/elinabian 1/Desktop/CU-life/intern/洋葱数学/videology_3.0/query_result.csv')
raw_data.columns = ['video_id', 'user_id', 'event_key', 'event_time', 
                    'drag_start', 'drag_end', 'time_stamp', 'answer_correct']
print(raw_data.head(20))
sorted_data = raw_data.sort_values('event_time')
sorted_data.index = pd.RangeIndex(len(sorted_data.index))
print(sorted_data.head(20))   #有特别多相同的event_time是否合理
```

```python
drag_opr = sorted_data[sorted_data['event_key']=='dragVideo']
start_opr = sorted_data[sorted_data['event_key']=='startVideo']
finish_opr = sorted_data[sorted_data['event_key']=='finishVideo']
exit_opr = sorted_data[sorted_data['event_key']=='clickVideoExit']

for i in drag_opr.index:
    a = drag_opr.loc[i, 'drag_start']/1000
    sorted_data.loc[i, 'drag_start'] = np.ceil(a)
    b = drag_opr.loc[i, 'drag_end']/1000
    sorted_data.loc[i, 'drag_end'] = np.ceil(b)
    if i%10000 == 0:
        print(i)
```

```python
for i in start_opr.index:
    c = start_opr.loc[i, 'time_stamp']/1000
    sorted_data.loc[i, 'time_stamp'] = np.ceil(c)
    if i%50000==0:
        print(i)
```

```python
for i in exit_opr.index:
    e = exit_opr.loc[i, 'time_stamp']/1000
    sorted_data.loc[i, 'time_stamp'] = np.ceil(e)
    if i%50000==0:
        print(i)
    
print(sorted_data.head(20))
```

```python
sorted_data = np.array(sorted_data)
length = sorted_data.shape[0]
print(length)
print(sorted_data[:10, :])
```

```python
extracted_data = []

for i in range(length):
    single_record = sorted_data[i,:]
    exist = False
    for ex_data in extracted_data:
        if ex_data['video_id'] == single_record[0] and ex_data['user_id'] == single_record[1]:
            if single_record[2] == 'startVideo':
                new_opr = {}
                new_opr['start_video'] = single_record[6]
                new_opr['forward'] = []
                new_opr['backward'] = []
                ex_data['operation'].append(new_opr)
            if single_record[2] == 'dragVideo':
                if single_record[4] < single_record[5]:
                    ex_data['operation'][-1]['forward'].append([single_record[4], single_record[5]])
                else:
                    ex_data['operation'][-1]['backward'].append([single_record[4], single_record[5]])
            if single_record[2] == 'finishVideo':
                ex_data['operation'][-1]['finish_video'] = single_record[6]
            if single_record[2] == 'clickVideoExit':
                ex_data['operation'][-1]['exit_video'] = single_record[6]
#            if single_record[2] == 'clickProblemSubmit':
#                ex_data['operation'][-1]['answer_question'] = single_record[7]
            exist = True
            break
    if exist == False and single_record[2] == 'startVideo':
        new_record = {}
        new_record['video_id'] = single_record[0]
        new_record['user_id'] = single_record[1]
        new_record['operation'] = []
        new_opr = {}
        new_opr['start_video'] = []
        new_opr['forward'] = []
        new_opr['backward'] = []
        new_opr['start_video'] = single_record[6]
        new_record['operation'].append(new_opr)
#        if single_record[2] == 'clickProblemSubmit':
#            new_opr['answer_question'] = single_record[7]
#            new_record['operation'].append(new_opr)
        extracted_data.append(new_record)
    if i%10000 == 0:
        print(i)
# print(extracted_data)
```

```python
client = pymongo.MongoClient(host='localhost', port=27017)
db = client['videology_bbbjy']
collection = db.videology_bbbjy
final = collection.insert_many(extracted_data)
```
