```python
import plotly.plotly as py
import plotly.graph_objs as go
import pickle
import pandas as pd
import numpy as np

# open file
with open('/Users/elinabian 1/Desktop/CU-life/intern/洋葱数学/exercise/watchratio.pkl', 'rb') as f:
    watch_ratio = pickle.load(f)
# print(watch_ratio['watch_ratio'])
with open('/Users/elinabian 1/Desktop/CU-life/intern/洋葱数学/exercise/drag_video_sample_result.pkl', 'rb') as f_2:
    drag_point = pickle.load(f_2)
# print(drag_point)
with open('/Users/elinabian 1/Desktop/CU-life/intern/洋葱数学/exercise/yuyun.pkl', 'rb') as f:
    miss_part = pickle.load(f)
# print(miss_part)



with open('/Users/elinabian 1/Desktop/CU-life/intern/洋葱数学/exercise/xiaoqing.pkl', 'rb') as f:
    aaa = pickle.load(f)

aaa = pd.DataFrame(aaa)
aaa.columns = ['s', 'e', 'd']
print(aaa)



# create watch ratio df
watch_ratio_df = pd.DataFrame(watch_ratio_df['watch_ratio'])
watch_ratio_df.columns = ['watch_ratio']
watch_ratio_df['time_line'] = range(0, len(watch_ratio_df))

# create drag point df
drag_point_df = pd.DataFrame(drag_point['dragZone'])
drag_point_df.columns = ['start_point', 'end_point', 'direction']

# create miss_part_df
miss_part_df = []
for i in range(len(miss_part)):
    aa = int(miss_part[i])
    miss_part_df.append(aa)
miss_part_df = pd.DataFrame(miss_part_df)
miss_part_df.columns = ['miss_part']

# calculate ratio mean
watch_ratio_mean = watch_ratio_df['watch_ratio'].mean()
mean_text = '{0:.2%}'.format(watch_ratio_mean)
mean_text = 'Average of Watch Ratio is ' + mean_text + '!'
print(mean_text)

#create drag interval assistant
drag_interval_text = []
for i in range(len(drag_point_df)):
    drag_interval_text.append('drag interval: from ' + str(drag_point_df.loc[i, 'start_point']) + ' to ' + str(drag_point_df.loc[i, 'end_point']) + 
                              '</br>drag direction: ' + str(drag_point_df.loc[i, 'direction']))
    
drag_line = []
for i in range(len(drag_point_df)):
    aa = list(range(drag_point_df.loc[i, 'start_point'], drag_point_df.loc[i, 'end_point']))
    drag_line.append(aa)
    
print(watch_ratio_df)
print(drag_point_df)
print(miss_part_df)
```

```python
# Create a trace
watch_ratio = go.Scatter(
    x=watch_ratio_df['time_line'],
    y=watch_ratio_df['watch_ratio'],
    name='watch ratio',
    mode='lines'
)

w_mean = go.Scatter(
    x=[len(watch_ratio_df)-25],
    y=[watch_ratio_mean-0.01],
    text=mean_text,
    mode='text',
    textfont=dict(
        color='rgb(150, 150, 150)',
        size=15,
        family='Arail',
    )
)

miss_part = go.Scatter(
    x=miss_part_df['miss_part'],
    y=[1] * len(miss_part_df),
    name='miss part',
    text='Students who didn\'t get the right answer miss these parts most!',
    mode='markers',
    marker = dict(
        color = 'red',
        symbol = 'star-dot',
        size = 12
    ))

drag_interval_1 = go.Scatter(
    x=drag_line[0],
    y=[1] * len(drag_line[0]),
    text = drag_interval_text[0],
    mode = 'markers',
    name = 'drag interval',
    marker = dict(
        color = 'rgba(255, 255, 255, 255)'
    ))

drag_interval_2 = go.Scatter(
    x=drag_line[1],
    y=[1] * len(drag_line[1]),
    text = drag_interval_text[1],
    mode = 'markers',
    name = 'drag interval',
    marker = dict(
        color = 'rgba(255, 255, 255, 255)'
    ))

drag_interval_3 = go.Scatter(
    x=drag_line[2],
    y=[1] * len(drag_line[2]),
    text = drag_interval_text[2],
    mode = 'markers',
    name = 'drag interval',
    marker = dict(
        color = 'rgba(255, 255, 255, 255)'
    ))

drag_interval_4 = go.Scatter(
    x=drag_line[3],
    y=[1] * len(drag_line[3]),
    text = drag_interval_text[3],
    mode = 'markers',
    name = 'drag interval',
    marker = dict(
        color = 'rgba(255, 255, 255, 255)'
    ))

drag_interval_5 = go.Scatter(
    x=drag_line[4],
    y=[1] * len(drag_line[4]),
    text = drag_interval_text[4],
    mode = 'markers',
    name = 'drag interval',
    marker = dict(
        color = 'rgba(255, 255, 255, 255)'
    ))

drag_interval_6 = go.Scatter(
    x=drag_line[5],
    y=[1] * len(drag_line[5]),
    text = drag_interval_text[5],
    mode = 'markers',
    name = 'drag interval',
    marker = dict(
        color = 'rgba(255, 255, 255, 255)'
    ))



layout = go.Layout(
    xaxis=dict(
        range=[-5, len(watch_ratio_df)+5],
        showline=True,
        showgrid=False,
        zeroline=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        autotick=True,
        ticks='outside',
        tickcolor='rgb(204, 204, 204)',
        tickwidth=2,
        ticklen=5,
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(150, 150, 150)',
        ),
    ),
    yaxis=dict(
        range=[watch_ratio_df['watch_ratio'].min()-0.05, watch_ratio_df['watch_ratio'].max()+0.05],
        title='Watch Ratio',
        titlefont=dict(
            size=20,
            color='rgb(150,150,150)'
        ),
        showgrid=False,
        showline=True,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        autotick=True,
        ticks='outside',
        tickcolor='rgb(204, 204, 204)',
        tickwidth=2,
        ticklen=5,
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(150, 150, 150)',
        ),
    ),
    autosize=False,
    showlegend=False,
    
)


annotations = []
annotations.append(dict(xref='paper', yref='paper', x=0.0, y=1.05,
                              xanchor='left', yanchor='bottom',
                              text='Analysis for the Teaching Video',
                              font=dict(family='Arial',
                                        size=30,
                                        color='rgba(49,130,189, 1)'),
                              showarrow=False))
annotations.append(dict(xref='paper', yref='paper', x=0.5, y=-0.1,
                              xanchor='center', yanchor='top',
                              text='Time Line for the Teaching Video',
                              font=dict(family='Arial',
                                        size=20,
                                        color='rgb(150,150,150)'),
                              showarrow=False))
for i in range(len(drag_point_df)):
    if drag_point_df.loc[i, 'direction'] == 'forward':
        annotations.append(dict(ax=drag_point_df['start_point'][i]-1, ay=0.85, axref='x', ayref='paper',
                                x=drag_point_df['end_point'][i]+1, y=0.85, xref='x', yref='paper',
                                showarrow=True, arrowhead=3, arrowsize=0.5, arrowwidth=4.5,
                                arrowcolor='rgba(255, 140, 164, 0.5)'
                               ))
    else:
        annotations.append(dict(ax=drag_point_df['end_point'][i]+1, ay=0.85, axref='x', ayref='paper',
                                x=drag_point_df['start_point'][i]-1, y=0.85, xref='x', yref='paper',
                                showarrow=True, arrowhead=3, arrowsize=0.5, arrowwidth=4.5,
                                arrowcolor='rgba(160, 64, 164, 0.5)'
                               ))


shapes = []
shapes.append(dict(type='line', xref='paper', yref='y',
                   x0=0, y0=watch_ratio_mean,
                   x1=1, y1=watch_ratio_mean,
                   line=dict(color='rgba(255, 140, 84, 0.5)',
                               width=4,
                               dash='dashdot'),))
for i in range(len(drag_point_df)):
    shapes.append(dict(type='rect', xref='x', yref='paper',
                       x0=drag_point_df['start_point'][i], y0=0, 
                       x1=drag_point_df['end_point'][i], y1=0.85,
                       fillcolor='#d3d3d3', opacity=0.2,
                       line=dict(width=0),),)


layout['annotations'] = annotations
layout['shapes'] = shapes
data = [watch_ratio, w_mean, drag_interval_1, drag_interval_2, drag_interval_3, drag_interval_4, drag_interval_5, drag_interval_6, miss_part]
fig = dict(data=data, layout=layout)

py.iplot(fig, filename='Videology_Plot')
```
